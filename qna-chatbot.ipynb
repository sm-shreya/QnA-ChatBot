```python
# Import all necessary libraries
from datasets import Dataset
from ragas import evaluate
from ragas.metrics import context_precision, context_recall
from ragas.llms import LangchainLLMWrapper

from langchain.chains import RetrievalQA
from langchain.retrievers import ContextualCompressionRetriever # Note: We will replace the Cohere-based reranker
from langchain.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.documents import Document
from langchain.load import dumps, loads

from langchain_groq import ChatGroq # <-- IMPORT GROQ
from langchain_cohere import CohereRerank # We still use Cohere for its high-quality reranker
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
print("✅ Step 1: Setup Complete.")


# --- 2. INITIALIZE BASE COMPONENTS (Economics PDF) ---
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

embedding_model = HuggingFaceEmbeddings(model_name="BAAI/bge-small-en-v1.5")

# Load your Economics PDF
loader = PyPDFLoader("/kaggle/input/eco-book/Chap01_Eco.pdf")
documents = loader.load()

# Split into smaller chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=300, chunk_overlap=150) 
chunked_docs = text_splitter.split_documents(documents)

# Build FAISS Vector Store from Economics content
vector_store = FAISS.from_documents(chunked_docs, embedding_model)
base_retriever = vector_store.as_retriever(search_kwargs={"k": 15})
print("✅ Step 2: Economics PDF Loaded and Vector Store Initialized.")



# --- 3. BUILD THE ADVANCED RAG-FUSION + RERANK RETRIEVER ---
# a. RAG-Fusion: Chain to generate multiple queries using Groq
template = """You are a helpful assistant that generates multiple search queries based on a single input query.
Generate 4 search queries, one on each line, related to the following input query:
Query: {original_query}
Queries:"""
prompt_rag_fusion = ChatPromptTemplate.from_template(template)
# Use a fast Groq model for query generation
groq_llm = ChatGroq(model_name="llama-3.1-8b-instant", temperature=0)

generate_queries = (
    prompt_rag_fusion
    | groq_llm
    | StrOutputParser()
    | (lambda x: x.strip().split("\n"))
)

# b. Chain to retrieve documents for each query and flatten the list
retrieval_chain = generate_queries | base_retriever.map() | (lambda x: [doc for sublist in x for doc in sublist])

# c. Cohere Reranker: We still use Cohere's excellent reranker for best results
# Note: You can skip this step for a 100% Groq-only pipeline, but the results will be better with it.
os.environ['COHERE_API_KEY'] = UserSecretsClient().get_secret("COHERE_API_KEY")
compressor = CohereRerank(model="rerank-english-v3.0", top_n=4)
advanced_retriever = ContextualCompressionRetriever(
    base_compressor=compressor, base_retriever=retrieval_chain
)
print("✅ Step 3: Advanced RAG-Fusion + Rerank Retriever Built (Query Generation by Groq).")


# --- 4. BUILD THE FINAL QA CHAIN USING GROQ ---
qa_chain = RetrievalQA.from_chain_type(
    llm=groq_llm, # <-- Use the fast Groq LLM for the final answer
    chain_type="stuff",
    retriever=advanced_retriever,
    return_source_documents=True
)
print("✅ Step 4: Final QA Chain Built with Groq.")


# --- 5. RUN THE RAGAS EVALUATION USING GROQ ---
eval_dataset = [
    {
        'question': "Briefly summarize the Economics PDF book.",
        'ground_truth': """The first chapter introduces economics as the study of scarcity, choice, 
opportunity cost, and the functioning of markets. It explains the division 
into microeconomics and macroeconomics, with examples such as airline 
pricing and consumer decisions."""
    },
    {
        "question": "What is economics?",
        "ground_truth": """Economics is the social science that studies how people, businesses, and governments make choices when resources are 
        limited (scarcity) to satisfy their needs and wants. It analyzes the production, distribution, and consumption of goods and services, 
        ranging from the decisions of individual consumers and firms (microeconomics) to the workings of entire economies and global trade 
        (macroeconomics). """
    },
    {
        "question": "What are some basic concepts in economics?",
        "ground_truth": """Basic economics concepts include scarcity (unlimited wants, limited resources), supply and demand (how prices are set), 
        opportunity cost (the value of the next best alternative given up), utility (satisfaction from consuming goods/services), and factors of 
        production (land, labor, capital, and entrepreneurship). Other key ideas are markets, equilibrium, elasticity (responsiveness to price changes), 
        and economic growth. Scarcity, opportunity cost, and market systems are fundamental concepts in economics."""
    }
]


questions = [ex['question'] for ex in eval_dataset]
ground_truths = [ex['ground_truth'] for ex in eval_dataset]

results = qa_chain.batch([{'query': q} for q in questions])

# Show outputs for inspection
for i, res in enumerate(results):
    print(f"\n🔎 Original Question: {questions[i]}")
    print(f"📝 Model Answer: {res['result']}")
    print(f"📖 Retrieved Contexts (first 200 chars each): {[doc.page_content[:200] for doc in res['source_documents']]}")

answers = [res['result'] for res in results]
contexts = [[doc.page_content for doc in res['source_documents']] for res in results]


response_dataset = Dataset.from_dict({
    "question": questions,
    "answer": answers,
    "contexts": contexts,
    "ground_truth": ground_truths
})

# Use the Groq model for the evaluation task itself
ragas_llm_eval = LangchainLLMWrapper(groq_llm)

evaluation_result = evaluate(
    dataset=response_dataset,
    metrics=[context_precision, context_recall],
    llm=ragas_llm_eval  # Pass the Groq wrapper
)
print("\n--- RAGAS Evaluation Results (Groq-Powered Pipeline) ---")
print(evaluation_result)
```
